Show me the formula that converts from Celsius to Fahrenheit
==============================

The formula to convert temperature from Celsius to Fahrenheit is:

F = [ (9/5) * C + 32 ]

where:
- \( F \) is the temperature in Fahrenheit,
- \( C \) is the temperature in Celsius.







Write me a program that has two vectors.
1. A vector named celcius contains a value from minus 10 to 10 in increments of 2.
2. Vector converts the values ​​of vector 1 to Fahrenheit.


```python
# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (9/5) * celsius + 32

# Vector of Celsius temperatures
celsius = list(range(-10, 12, 2))

# Convert Celsius to Fahrenheit and store in a new vector
fahrenheit = [celsius_to_fahrenheit(temp) for temp in celsius]

# Print the original and converted temperature vectors
print("Celsius:", celsius)
print("Fahrenheit:", fahrenheit)
```
